```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("camembert.csv", header=0, sep=",")

# Extraire les colonnes
categories = df['Categories']
votes = df['Votes']

# Crer le graphique  barres
plt.figure(figsize=(12,6))
plt.bar(categories, votes, color='skyblue')
plt.ylabel('Nombres de votes')
plt.title('Rpartition des votes par catgorie')
plt.xticks(rotation=45)
plt.tight_layout()  
plt.show()

```
